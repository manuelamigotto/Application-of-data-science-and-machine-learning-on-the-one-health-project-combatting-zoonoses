This notebook is for create files with geocoding (latitude, longitude).<br>
<b>STEPS:</b>
<br>
<ol>
  <li>Divide by region datafreme with all records.</li>
  <li>Divide by disease each file generated from previous step.</li>
  <li>Geocode.</li>
</ol>

```python
import pandas as pd
from split_utils import split_by_region_to_csv
import pyarrow.feather as feather
from split_utils import split_by_zoonoses
from geocode_utils import create_complete_address, create_rag_soc_address, create_centroid_address
from geocode import massiveGoogleGeocode
from geocode_nominatim import massiveNominatimGeocode,get_notFound

prefix_in_out_file = 'dataframe_sigla'#'dataframe_sigla_water_others'
target_region = ["Piemonte","VDA","Liguria","Sicilia","Marche","Abruzzo","Toscana","Campania","Puglia","Veneto","Lombardia","EmiliaRomagna","TAA","Sardegna","Molise","Calabria","Lazio","FVG","Basilicata","Umbria"]#"Piemonte","VDA","Liguria","Sicilia","Marche","Abruzzo","Toscana","Campania","Puglia","Veneto","Lombardia","EmiliaRomagna","TAA","Sardegna","Molise","Calabria","Lazio","FVG","Basilicata","Umbria"
diseases = ["epatite","leishmania","leptospira","salmonella"] #"epatite","leishmania","leptospira","salmonella"
json_addr_dict = 'distances_features_2_5.json'
```

```python
#STEP 1
for region in target_region:
    split_by_region_to_csv(prefix_in_out_file, region, "feather")
```

```python
#STEP 2
split_by_zoonoses(prefix_in_out_file, target_region)
```

```python
#STEP 3
for region in target_region:
    for disease in diseases:
        file_input = region + "/" + prefix_in_out_file + "_" + region + "_" + disease + ".csv" #"df_sigla_geocoded_" + region + ".csv"
        file_output = region + "/" + prefix_in_out_file + "_" + region + "_" + disease + "_output.csv"
        
        #retrive row_id of rows with geocoding out of their region
        create_complete_address(file_input, file_output,"csv",printOn=False)
        massiveNominatimGeocode(region, file_output, file_output, json_addr_dict)
        continue_geocoding = massiveGoogleGeocode(region, file_output, file_output, json_addr_dict)
        num_geocode = continue_geocoding[0]
        num_addresses = continue_geocoding[1]
        if num_addresses != num_geocode:
            create_rag_soc_address(file_output,file_output, "csv")
            continue_geocoding = massiveGoogleGeocode(region, file_output, file_output, json_addr_dict)
            num_geocode += continue_geocoding[0]
            if num_addresses != num_geocode:
                create_centroid_address(file_output,file_output, "csv")
                massiveNominatimGeocode(region, file_output, file_output, json_addr_dict, True)
                not_found_remaining = get_notFound(file_output)
                if not_found_remaining:
                    print("Attention! There are addresses (as centroid) not geocoded!")               
```
